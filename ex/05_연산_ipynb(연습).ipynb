### 5.2. ÏãúÎ¶¨Ï¶àÏùò Ïó∞ÏÇ∞

üìå ÏãúÎ¶¨Ï¶à Ïó∞ÏÇ∞
- **Îã®Ïùº Í∞íÍ≥º Ïó∞ÏÇ∞**: Î™®Îì† ÏõêÏÜåÏóê ÎèôÏùº Ï†ÅÏö©
- **ÏãúÎ¶¨Ï¶à Í∞ÑÏùò Ïó∞ÏÇ∞**: Ïù∏Îç±Ïä§ Í∏∞Ï§Ä ÏÑ±Î∂ÑÎ≥Ñ Ïó∞ÏÇ∞  

#### 5.2.1. ÏãúÎ¶¨Ï¶àÏôÄ Îã®Ïùº Í∞íÏùò Ïó∞ÏÇ∞

- Î™®Îì† ÏõêÏÜåÏóê ÎèôÏùº Ï†ÅÏö©

```python
# ÏΩîÎìú 5-1. Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
import pandas as pd
s1 = pd.Series([1, 5, 0, 4], index=list('ABCD'))
s2 = pd.Series([1, 2, 3, 3], index=list('ABCD'))
s3 = pd.Series([3, 4, 2, 3], index=list('ABDE'))
```

```python
s1
```

```python
s2
```

```python
# ÏΩîÎìú 5-2. ÏãúÎ¶¨Ï¶àÏôÄ Îã®Ïùº Í∞íÏùò ÏÇ∞Ïà† Ïó∞ÏÇ∞ ÎçßÏÖà
# s1Í≥º 2Ïùò ÎçßÏÖà(Ïó∞ÏÇ∞Ïûê)
s1+2
# s1Í≥º 2Ïùò ÎçßÏÖà(Ïó∞ÏÇ∞ Ìï®Ïàò)
s1.add(2)
```

```python
# ÏΩîÎìú 5-3. ÏãúÎ¶¨Ï¶àÏôÄ Îã®Ïùº Í∞íÏùò ÎπÑÍµê Ïó∞ÏÇ∞
# s1Í≥º 3ÏùÑ ÎπÑÍµêÌïòÏó¨, s1Ïù¥ 3Î≥¥Îã§ ÌÅ∞ÏßÄ ÌôïÏù∏(Ïó∞ÏÇ∞Ïûê)
s1>3

# s1Í≥º 3ÏùÑ ÎπÑÍµêÌïòÏó¨, s1Ïù¥ 3Î≥¥Îã§ ÌÅ∞ÏßÄ ÌôïÏù∏ (Ïó∞ÏÇ∞ Ìï®Ïàò)
s1.gt(3)
```

#### 5.2.2. ÏãúÎ¶¨Ï¶à Í∞ÑÏùò Ïó∞ÏÇ∞
- Ïù∏Îç±Ïä§ Í∏∞Ï§Ä ÏÑ±Î∂ÑÎ≥Ñ Ïó∞ÏÇ∞

```python
# ÏΩîÎìú 5-4. Ïù∏Îç±Ïä§Í∞Ä ÏùºÏπòÌïòÎäî Îëê ÏãúÎ¶¨Ï¶àÏùò Î∫ÑÏÖà
# s1Í≥º s2Ïùò Î∫ÑÏÖà(Ïó∞ÏÇ∞Ïûê)
s1-s2
# s1Í≥º s2Ïùò Î∫ÑÏÖà(Ïó∞ÏÇ∞ Ìï®Ïàò)
s1.sub(s2)
```

```python
# ÏΩîÎìú 5-5. Ïù∏Îç±Ïä§Í∞Ä ÏùºÏπòÌïòÎäî Îëê ÏãúÎ¶¨Ï¶àÏùò ÎèôÎì± ÎπÑÍµê Ïó∞ÏÇ∞(==)
# s1Í≥º s2Ïùò ÎèôÎì± ÎπÑÍµê(Ïó∞ÏÇ∞Ïûê)
s1==s2

# s1Í≥º s2Ïùò ÎèôÎì± ÎπÑÍµê(Ïó∞ÏÇ∞ Ìï®Ïàò)
s1.eq(s2)
```

```python
# ÏΩîÎìú 5-6. ÏãúÎ¶¨Ï¶àÏùò ÎÖºÎ¶¨ Ïó∞ÏÇ∞
# s1ÏóêÏÑú Í∞íÏù¥ 0Î≥¥Îã§ ÌÅ¨Í≥† 3Î≥¥Îã§ ÏûëÏùÄ ÏõêÏÜå Ï∂îÏ∂ú (Ïó∞ÏÇ∞Ïûê)
(s1 > 0) & (s1 < 3)
# s1ÏóêÏÑú Í∞íÏù¥ 0Î≥¥Îã§ ÌÅ¨Í≥† 3Î≥¥Îã§ ÏûëÏùÄ ÏõêÏÜå Ï∂îÏ∂ú (Ïó∞ÏÇ∞ Ìï®Ïàò)
s1.gt(0)&s1.lt(3)
```

```python
# ÏΩîÎìú 5-7. Ïù∏Îç±Ïä§Ïùò Íµ¨ÏÑ±Ïù¥ Îã§Î•∏ Îëê ÏãúÎ¶¨Ï¶àÏùò Í≥±ÏÖà
# s1Í≥º s3Ïùò Í≥±ÏÖà(Ïó∞ÏÇ∞Ïûê)
s1*s3
# s1Í≥º s3Ïùò Í≥±ÏÖà(Ïó∞ÏÇ∞ Ìï®Ïàò)
s1.mul(s3)
```

```python
# ÏΩîÎìú 5-8. Ïù∏Îç±Ïä§Ïùò Íµ¨ÏÑ±Ïù¥ Îã§Î•∏ Îëê ÏãúÎ¶¨Ï¶àÏùò ÎπÑÍµê Ïó∞ÏÇ∞(Î∞òÎìúÏãú Ïó∞ÏÇ∞ Ìï®Ïàò)
# s1Ïùò Í∞Å ÏõêÏÜåÍ∞Ä s3Ïùò ÎåÄÏùë ÏõêÏÜåÎ≥¥Îã§ ÏûëÍ±∞ÎÇò Í∞ôÏùÄÏßÄ ÎπÑÍµê (<= Ïó∞ÏÇ∞)
#s1 <= s3 ÏùÄ ÏóêÎü¨Í∞Ä ÎÇòÏò®Îã§
s1.le(s3)
```

#### 5.2.2. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏóêÏÑú Ïó¥ Í∞ÑÏùò Ïó∞ÏÇ∞

üìå Ìïú Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÎÇ¥Î∂ÄÏóêÏÑú Ïó¥ Í∞ÑÏùò Ïó∞ÏÇ∞

- Í∞Å Ïó¥ÏùÄ Ïù∏Îç±Ïä§Í∞Ä ÎèôÏùºÌïú **ÏãúÎ¶¨Ï¶à(Series)**  
- Ïó¥ Í∞Ñ Ïó∞ÏÇ∞ = **ÎèôÏùºÌïú ÏúÑÏπòÏùò ÏÑ±Î∂Ñ Ïó∞ÏÇ∞**  

```python
# ÏΩîÎìú 5-9. Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
import pandas as pd
data1 = {'ÏÑ±': ['ÍπÄ', 'Í∞ï', 'Í∂å', 'Î∞ï'],
         'Ïù¥Î¶Ñ': ['ÌåêÎã§', 'ÏäπÏ£º', 'Î≥¥ÏïÑ', 'Ìö®Ïã†'],
         'Íµ≠Ïñ¥': [79, 65, 92, 95],
         'ÏòÅÏñ¥': [68, 83, 87, 79]}
df = pd.DataFrame(data1)
df
```

```python
# ÏΩîÎìú 5-10. dfÏóêÏÑú Íµ≠Ïñ¥ Ïó¥Í≥º ÏòÅÏñ¥ Ïó¥ÏùÑ ÎçîÌï¥ Ï¥ùÏ†ê Ïó¥ ÏÉùÏÑ±ÌïòÍ∏∞

df['Íµ≠Ïñ¥']
df['ÏòÅÏñ¥']
df['Ï¥ùÏ†ê'] = df['Íµ≠Ïñ¥'] + df['ÏòÅÏñ¥']
df
```

```python
# ÏΩîÎìú 5-11. dfÏóêÏÑú Ï¥ùÏ†ê Ïó¥ÏùÑ 2Î°ú ÎÇòÎàÑÏñ¥ ÌèâÍ∑† Ïó¥ ÏÉùÏÑ±ÌïòÍ∏∞

df['ÌèâÍ∑†']=df['Ï¥ùÏ†ê']/2
df
```

```python
# ÏΩîÎìú 5-12. dfÏóêÏÑú Íµ≠Ïñ¥ Ï†êÏàòÍ∞Ä ÏòÅÏñ¥ Ï†êÏàòÎ≥¥Îã§ ÎÜíÏùÄ ÏÇ¨Îûå ÌôïÏù∏ÌïòÍ∏∞
df['Íµ≠Ïñ¥']>df['ÏòÅÏñ¥']
```

```python
# ÏΩîÎìú 5-13. dfÏóêÏÑú ÌèâÍ∑† Ï†êÏàòÍ∞Ä 80Ï†ê ÎØ∏ÎßåÏù∏ ÏÇ¨Îûå ÌôïÏù∏ÌïòÍ∏∞
df['ÌèâÍ∑†'] < 80
```

```python
# ÏΩîÎìú 5-14. dfÏóêÏÑú Íµ≠Ïñ¥ Ï†êÏàòÍ∞Ä ÏòÅÏñ¥ Ï†êÏàòÎ≥¥Îã§ ÎÜíÏúºÎ©¥ÏÑú, 80Ï†ê ÎØ∏ÎßåÏù∏ ÏÇ¨Îûå ÌôïÏù∏ÌïòÍ∏∞
(df['Íµ≠Ïñ¥'] > df['ÏòÅÏñ¥'])&(df['Íµ≠Ïñ¥'] < 80)
#ÌòπÏùÄ df['Íµ≠Ïñ¥'].gt(df['ÏòÅÏñ¥'])&(df['Íµ≠Ïñ¥'] < 80)
```

```python
# ÏΩîÎìú 5-15. dfÏóêÏÑú ÏÑ± Ïó¥Í≥º Í≥µÎ∞±, Ïù¥Î¶Ñ Ïó¥ÏùÑ Ìï©Ï≥ê ÏÑ±Î™Ö Ïó¥ÏùÑ ÏÉùÏÑ±
df['ÏÑ±Î™Ö']= df['ÏÑ±'] + ' ' + df['Ïù¥Î¶Ñ']
# ÏÑ±Î™Ö, Íµ≠Ïñ¥, ÏòÅÏñ¥, ÌèâÍ∑† Ïó¥Îßå Ï∂úÎ†•ÌïòÍ∏∞
df[['ÏÑ±Î™Ö','Íµ≠Ïñ¥','ÏòÅÏñ¥','ÌèâÍ∑†']]
```

### 5.3. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùò Ïó∞ÏÇ∞

- **Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Í∞Ñ Ïó∞ÏÇ∞**: Ïù∏Îç±Ïä§¬∑Ïª¨ÎüºÏ¶à Í∏∞Ï§ÄÏúºÎ°ú ÎèôÏùº ÏúÑÏπò ÏÑ±Î∂ÑÎÅºÎ¶¨ Ïó∞ÏÇ∞

#### 5.3.1. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÍ≥º Îã®Ïùº Í∞íÏùò Ïó∞ÏÇ∞
- Î™®Îì† ÏÖÄÏóê ÎèôÏùº Ï†ÅÏö©

```python
# ÏΩîÎìú 5-16. Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
import pandas as pd
data1 = [[0, 1, 1], [0, 2, 1], [4, 2, 5], [2, 2, 5]]
data2 = [[3, 2, 1], [3, 4, 5], [0, 3, 4], [0, 0, 4]]
df1 = pd.DataFrame(data1, index=list('abcd'), columns=list('ABC'))
df2 = pd.DataFrame(data2, index=list('abcd'), columns=list('ABC'))
df3 = pd.DataFrame(data2, index=list('abde'), columns=list('ABD'))
```

```python
df1
```

```python
# ÏΩîÎìú 5-17. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÍ≥º Îã®Ïùº Í∞íÏùò ÏÇ∞Ïà† Ïó∞ÏÇ∞(Î™´)
# df1ÏùÑ 3ÏúºÎ°ú ÎÇòÎàà Î™´(Ïó∞ÏÇ∞Ïûê)
df1//3
# df1ÏùÑ 3ÏúºÎ°ú ÎÇòÎàà Î™´(Ïó∞ÏÇ∞ Ìï®Ïàò)
df1.floordiv(3)
```

```python
# ÏΩîÎìú 5-18. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÍ≥º Îã®Ïùº Í∞íÏùò ÎπÑÍµê Ïó∞ÏÇ∞(Îã§Î•¥Îã§)
# df1Ïùò Í∞Å ÏõêÏÜåÍ∞Ä 2ÏôÄ Îã§Î•¥Îã§(!= ) Ïó∞ÏÇ∞(Ïó∞ÏÇ∞Ïûê)
df1 != 2
# df1Ïùò Í∞Å ÏõêÏÜåÍ∞Ä 2ÏôÄ Îã§Î•¥Îã§(!= ) Ïó∞ÏÇ∞(Ïó∞ÏÇ∞ Ìï®Ïàò)
df1.ne(2)
```

```python
# ÏΩîÎìú 5-19. Ïù∏Îç±Ïä§ÏôÄ Ïª¨ÎüºÏ¶àÍ∞Ä ÏùºÏπòÌïòÎäî Îëê Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùò ÏÇ∞Ïà† Ïó∞ÏÇ∞
# df1Í≥º df2Ïùò ÎçßÏÖà(Ïó∞ÏÇ∞Ïûê)
df1+df2
# df1Í≥º df2Ïùò ÎçßÏÖà(Ïó∞ÏÇ∞ Ìï®Ïàò)
df1.add(df2)
```

#### 5.3.3. NaNÏùÑ ÎåÄÏ≤¥Ìï¥ Ïó∞ÏÇ∞ÌïòÍ∏∞(Ïó∞ÏÇ∞ Ìï®ÏàòÏùò fill_value)

- **Í≤∞Í≥º**: ÏãúÎ¶¨Ï¶àÍ∞Ä Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÌÅ¨Í∏∞ÎßåÌÅº ÌôïÏû•Îêú Îí§ Ïó∞ÏÇ∞Ìïú Í≤ÉÍ≥º ÎèôÏùº

```python
# ÏΩîÎìú 5-23. Ïã§Ïäµ ÏòàÏ†ú ÏΩîÎìú
import pandas as pd
data1 = [[1, 1, 1], [1, 2, 3], [4, 3, 5], [2, 2, 4]]
df = pd.DataFrame(data1, index=list('abcd'), columns=list('ABC'))
s1 = pd.Series([0, 1, 2], index=list('ABC'))
s2 = pd.Series([0, 1, 2, 3], index=list('abcd'))
```

### 5.4. ÌÜµÍ≥Ñ Ìï®Ïàò Ï†ÅÏö©ÌïòÍ∏∞

| Ìï®Ïàò | Í∏∞Îä• | Ìï®Ïàò | Í∏∞Îä• |
| :--- | :--- | :--- | :--- |
| **`sum`** | Ï†ÑÏ≤¥ Ìï© | **`cumsum`** | ÎàÑÏ†Å Ìï© |
| **`prod`** | Ï†ÑÏ≤¥ Í≥± | **`cumprod`** | ÎàÑÏ†Å Í≥± |
| **`max`** | ÏµúÎåìÍ∞í | **`cummax`** | ÎàÑÏ†Å ÏµúÎåÄ |
| **`min`** | ÏµúÏÜüÍ∞í | **`cummin`** | ÎàÑÏ†Å ÏµúÏÜå |
| **`count`** | Í∞úÏàò | **`cumcount`** | ÎàÑÏ†Å Í∞úÏàò(Í∑∏Î£πÌôî) |
| **`mean`** | ÌèâÍ∑† | **`first`** | Ï≤´ Î≤àÏß∏ Í∞í(Í∑∏Î£πÌôî) |
| **`std`** | ÌëúÏ§ÄÌé∏Ï∞® | **`last`** | ÎßàÏßÄÎßâ Í∞í(Í∑∏Î£πÌôî) |
| **`var`** | Î∂ÑÏÇ∞ | **`all`** | Î™®Îëê TrueÎùºÎ©¥ True(Î∂à) |
| **`median`** | Ï§ëÍ∞ÑÍ∞í | **`any`** | ÌïòÎÇòÎùºÎèÑ TrueÎùºÎ©¥ True(Î∂à) |
| **`quantile`** | Î∂ÑÏúÑÏàò | **`idxmax`** | ÏµúÎåìÍ∞íÏùò ÏÉâÏù∏ |

### üí° Ï£ºÏöî Ìï®Ïàò Ï∞∏Í≥† ÏÇ¨Ìï≠

- **`cumcount`**, **`first`**, **`last`**: Îã®ÎèÖÏúºÎ°ú ÏÇ¨Ïö©Ìï† Ïàò ÏóÜÍ≥†, Î∞òÎìúÏãú **`groupby`**, **`pivot_table`**, **`resample`** Îì± Í∑∏Î£πÌôî(Aggregation)Î•º ÏàòÌñâÌïòÎäî Ìï®ÏàòÏôÄ Ìï®Íªò ÏÇ¨Ïö©.

- **`idxmax` / `idxmin`**: ÏµúÎåÄÍ∞í ÎòêÎäî ÏµúÏÜåÍ∞í ÏûêÏ≤¥Í∞Ä ÏïÑÎãàÎùº, Ìï¥Îãπ Í∞íÏù¥ ÏúÑÏπòÌïú Ïù∏Îç±Ïä§Î•º Î∞òÌôòÌïòÎ©∞ **ÏµúÎåÄÎÇò ÏµúÏÜåÍ∞Ä Î≥µÏàòÏùº ÎïåÎäî Ï≤´ Î≤àÏß∏ Í∞íÏùò Ïù∏Îç±Ïä§Î•º Î∞òÌôò**

- **`all` / `any`**: Ï£ºÎ°ú Î∂àÎ¶¨Ïñ∏ Ïù∏Îç±Ïã± Í≤∞Í≥ºÏóêÏÑú Î™®Îì† Ï°∞Í±¥Ïù¥ Ï∂©Ï°±ÎêòÏóàÎäîÏßÄ ÌòπÏùÄ ÌïòÎÇòÎùºÎèÑ Ï∂©Ï°±ÎêòÏóàÎäîÏßÄ ÌôïÏù∏Ìï† Îïå ÏÇ¨Ïö©

- **`axis` ÏÑ§Ï†ï**: ÏúÑ Ìï®ÏàòÎì§ÏùÄ Í∏∞Î≥∏Ï†ÅÏúºÎ°ú `axis=0`(Ïó¥ Îã®ÏúÑ)ÏúºÎ°ú ÏûëÎèôÌïòÎ©∞, Ìñâ Îã®ÏúÑ Í≥ÑÏÇ∞Ïù¥ ÌïÑÏöîÌï† Í≤ΩÏö∞ `axis=1`ÏùÑ ÏßÄÏ†ï
