### 3.3. 파일에서 데이터 프레임 불러오기

#### 3.3.1. 엑셀 파일에서 데이터 프레임 불러오기(read_excel)

````python
# 코드 3-17. 엑셀 파일의 첫 번째 시트를 데이터 프레임으로 불러오기
import pandas as pd
url1 = 'https://github.com/panda-kim/book1/blob/main/01re.xlsx?raw=true'
df = pd.read_excel(url1)
df
````

````python
# 코드 3-18. 엑셀 파일의 두 번째 시트를 데이터 프레임으로 불러오기
df = pd.read_excel(url1, sheet_name='sheet_name')
#
df
````

````python
# 코드 3-19. 시트의 로케이션으로 두 번째 시트 불러오기
df = pd.read_excel(url1, sheet_name=1)
#시트 이름 대신에 시트의 순서대로 0,1,2,3 세어서sheet_name=1 <-이렇게 표현해도 됨
df
````

````python
# 코드 3-20. 세 번째 행을 컬럼즈로 지정해 세 번째 시트 불러오기
df = pd.read_excel(url1, sheet_name=2, header=2)
df
````

````python
# 코드 3-21. 맨 왼쪽 열을 인덱스로 지정해 네 번째 시트 불러오기
df = pd.read_excel(url1, sheet_name=3, index_col=0)
df
````

````python
# 코드 3-22. 다섯 번째 시트를 불러 오면서 학생코드 열을 문자열로 지정
#학생코드를 숫자가 아닌 문자열로 하고 싶을 때 사용하는 것 -> dtype={'학생코드':'str'}
df = pd.read_excel(url1, sheet_name=4, dtype={'학생코드':'str'})
df
````

#### 3.3.2. CSV 파일에서 데이터 프레임 불러오기(read_csv)


````python
# 코드 3-23. CSV 파일에서 데이터 프레임 불러오기
#csv 파일은 콤마로 데이터를 구분한다. 장점: 빅데이터를 다룰 떄 용량이 적은 이점
url1 = 'https://github.com/panda-kim/book1/blob/main/02read_csv1.csv?raw=true'
df = pd.read_csv(url1)
df
````
